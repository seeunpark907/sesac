# 3. 응답 생성 방식

API를 호출하여 모델의 응답을 생성하는 두 가지 방식이 있음
- invoke(): 모든 토큰이 생성된 후 하나의 string객체로 일괄적으로 반환하는 방식
- stream(): 생성되는 토큰을 즉시 반환하는 방식

## 3.1 invoke()

```python
from dotenv import load_dotenv
load_dotenv()
```

```python
# 아래의 question으로 직접 모델 객체를 생성하여 모델의 응답 받아보기
from langchain_openai import ChatOpenAI
  
question = "서울의 수도는 어디야?"

model = ChatOpenAI(model="gpt-5-nano")  # 모델객체 생성의 의미는 채팅방 개설!(채팅방에서 사용할 모델 지정)
response = model.invoke(question)

print(response.content)
```

## 3.2 stream()

```python
# from langchain_openai import ChatOpenAI

model = ChatOpenAI(model="gpt-5-nano")
question = "나는 영어영문학을 전공했는데, 데이터분석가로 직무전환을 하려해. 데이터 분석가가 되기 위한 과정을 알려줘."

print(model.stream(question))  
# invoke는 바로 반환되지만 stream은 반환할 요소들을 메모리에 올려두었다는 generator 객체만 반환. generator객체는 반복문을 통해서 요소를 반환해야함.

# generator 객체는 당장 생성되는 값에 대해서만 메모리공간을 차지
# 값이 반환되는 즉시 해당 공간이 비워지기 때문에 메모리 공간을 효율적으로 쓸 수 있음
# 따라서 generator객체는 print()문을 통해 호출하더라도 값을 반환할 준비상태를 출력해주고 원하는 결과를 출력해주지 않는다.
# 원하는 결과를 콘솔에서 확인하려면 for문을 통해 확인가능
```

```python
# token 형태 확인
for token in model.stream(question):
    print(token)  # 하나의 content라는 객체 안에 한개의 토큰씩 반환됨 
```

```python
# token 객체에서 content 속성만 출력해보기
for token in model.stream(question):
    print(token.content)  # token객체에서 content 속상만 출력
```

```python
# print()문에 end옵션을 활용하여 읽기 편하게 출력하기
for token in model.stream(question):
    print(token.content, end="", flush=False) # 결과값이 변수에 담기지 않았기떄문에 출력된 결과값을 새로운 셀에서 재활용 불가!
    # flush=True: 리소스 소모가 커져서 성능 저하 가능성 있음. (버퍼 없는 빈번한 I(input)/O(output)작업은 cpu 사용량이 증가할 수 있음) false 설정을 추천.
```

```python
# 생성되는 token을 변수에 할당하여 응답을 재활용하기
response = "" 
for token in model.stream(question):
    response += token.content
    print(token.content, end="", flush=True)
```

# 4. 모델의 응답 방식 제어

- temperature:
    - 0~2 사이의 값으로 설정 (2에 가까울수록 ai의 창의성을 높여줌)
    - 값이 높을수록 출력이 더 무작위적으로 생성되며, 값이 낮을수록 출력이 더 결정론적으로 생성됨 
    - 즉, 모델의 창의성을 제어하는 설정값으로 0에 가까울수록 일관되고 정확한 답변이 생성됨

```python
# from langchain_openai import ChatOpenAI

# temperature=0으로 설정할 경우
model_temp_0 = ChatOpenAI(
    model="gpt-5-nano",
    temperature=0,
)

response_temp_0 = model_temp_0.invoke("왜 하늘은 하늘색인가요?")
print(response_temp_0.content)
```

```python
# from langchain_openai import ChatOpenAI

# temperature=2로 설정할 경우
model_temp_2 = ChatOpenAI(
    model="gpt-5-nano",
    temperature=2,  # temperature=2는 선택할수있는 토큰의 범위를 확넓혀서 말도 안되는 답변을 하는데 5nano모델이 성능이 좋아서 답변이 잘나옴.
)

response_temp_2 = model_temp_2.invoke("왜 하늘은 하늘색인가요?")
print(response_temp_2.content)
```

# 5. Role Base 프롬프트 작성 방식

- 프롬프트 구성: System, User, AI

    - **System Prompt**:
        - 개발할 서비스의 전역적인 설정을 위한 프롬프트
        - 모델을 통해 어떤 서비스를 개발한건지 또는 모델에 어떤 페르소나를 적용시키고 응답은 어떠한 포맷으로 생성할건지 등을 설정하는 프롬프트.

    - **User Prompt**:
        - 개발할 서비스의 지역적인 설정을 위한 프롬프트로
        - 즉, 현재 작성중인 명령 또는 질문에만 해당하는 응답을 설정하기 위한 프롬프트
        - 개발할 서비스를 사용할 사용자의 질의 및 명령이 전달되는 프롬프트
            - 개발자의 명령과 서비스를 사용할 사용자의 명령이 함께 작성될 수도 있고 또는 사용자의 명령만 작성될 수도 있음 

    - **AI Prompt**:
        - 모델의 응답으로, 모델을 학습시키는 경우 원하는 형태의 응답 포맷을 직접 작성하여 학습시킬 수 있음(지도학습과 비슷한 맥락)

## 5.1 User Prompt - 사용자의 질의 및 명령만 전달할 경우

서비스를 사용할 사용자의 질의 및 명령만 User 프롬프트에 작성되어 모델에 전달되는 경우

```python
from langchain_openai import ChatOpenAI

model = ChatOpenAI(model="gpt-5-nano")

# 서비스를 사용할 사용자(고객)의 질의
question = "최근 카드값이 너무 많이 나와서 걱정이에요"  

response = model.invoke(question) # User프롬프트에 input data(고객의 질의)만 작성되어 전달된 경우. 
print(response.content)

# input-data만 전달될 경우 매 번 모델의 응답이 달라질 수 있음
# 일관되고 정확한 답변만 생성하기 위해서는 추가적인 지시사항이 필요함(프롬프트 엔지니어의 역할)
```

## 5.2 User Prompt - 개발자의 명령과 사용자의 질의가 함께 전달된 경우

User 프롬프트에 개발자의 지역적 명령과 서비스를 사용할 고객의 질의가 함께 작성되어 모델에 전달될 경우

```python
# from langchain_openai import ChatOpenAI
# model = ChatOpenAI(model="gpt-5-nano")

# 서비스를 사용할 사용자(고객)의 질의 
question = "최근 카드값이 너무 많이 나와서 걱정이에요"  
# question은 고객의 질의이기 떄문에 매번 달라지니까 f스트링 값 안에서 변수로 전달하고 지시사항은 매번 일정하기 때문에 별도로 전달.

user_prompt = f"""\
아래 고객의 금융 문제를 분석하여 가능성 있는 원인 3가지와 각 원인에 따른 실질적인 해결 방법을 제시해주세요.

--사용자 질의--
{question}\
"""

response = model.invoke(user_prompt)
print(response.content)
```

## 5.3 System Prompt

system prompt와 user prompt를 각각 작성하여 모델에 전달할 경우

### 5.3.1 예제1 - 금융

```python
system_prompt = """\
당신은 은행의 금융 상담 전문가입니다. 

상담 원칙:
1. 고객의 금융 문제를 정확히 파악합니다.
2. 3가지 가능한 원인과 각각의 구체적인 해결방안을 제시합니다.
3. 금융 용어는 쉽게 풀어서 설명합니다. 
4. 실행 가능한 단계별 액션 플랜을 제공합니다.
5. 필요시 금융 상품이나 서비스를 추천합니다.

출력 포맷:
- 문제 상황 요약
- 원인 1: [원인 설명] -> 해결 방법: [구체적인 실행 방안]
- 원인 2: [원인 설명] -> 해결 방법: [구체적인 실행 방안]
- 원인 3: [원인 설명] -> 해결 방법: [구체적인 실행 방안]
- 추가 조언 및 추천 서비스\
"""

question = "최근 카드값이 너무 많이 나와서 걱정이에요"

user_prompt = f"""\
아래 고객의 금융 문제를 분석하여 가능성 있는 원인 3가지와 각 원인에 따른 실질적인 해결방법을 제시해주세요.

--고객 문의--
{question}
"""
```

```python
# from langchain_openai import ChatOpenAI
# model = ChatOpenAI(model="gpt-5-nano")

# messages 객체 생성:
# langchain에서는 openai inerface를 지원하기 때문에 openai와 작성방식을 일치시켜도 무방
messages = [
    {
        "role": "system",
        "content": system_prompt
    },
    {
        "role": "user",
        "content": user_prompt
    }
]

# 모델 응답 받기
response = model.invoke(messages)
print(response.content)
```
