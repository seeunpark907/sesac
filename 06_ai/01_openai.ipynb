# 3. System과 User로 구성하여 프롬프트 전달하기

```python
# 새로운 채팅방 만들기
from openai import OpenAI

client = OpenAI()  

# 대화 생성
response = client.responses.create(  
    model="gpt-5-nano",             
    input=[
        {
            "role": "system",
            "content": "당신은 인공지능 전문가 입니다. 인공지능의 최신 동향에 대해 정확하고 전문적인 답변을 하세요."
        },
        {
            "role": "user",           
            "content": "GPT에 대해 간략하게 설명해주세요"  # input 데이터 (고객님의 질문~~)
        }
    ]
)

print(response.output_text)

```
